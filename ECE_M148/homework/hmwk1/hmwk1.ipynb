```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# ensures that we are inline plots
```

```python
from sklearn.neighbors import KNeighborsRegressor
# to use KNN
```

```python
X = np.array([1, 2, 3.2, 4, 5, 6])
y = np.array([1, 4, 6, 3, 2, 2])
X = X.reshape(-1, 1)
```

```python
# plotting the plots first
plt.scatter(X, y, color = "red")
plt.grid()
```

```python
# K = 1
KNN1 = KNeighborsRegressor(n_neighbors = 1)
KNN1.fit(X, y)
y_est1 = KNN1.predict(np.arange(0, 8, .01).reshape(-1, 1))
```

```python
plt.scatter(X, y, color = "red", label = "K = 1")
plt.plot(np.arange(0, 8, .01), y_est1)
plt.grid()
plt.show()
```

```python
# K = 2
KNN2 = KNeighborsRegressor(n_neighbors = 2)
KNN2.fit(X, y)
y_est2 = KNN2.predict(np.arange(0, 8, .01).reshape(-1, 1))
```

```python
plt.scatter(X, y, color = "red", label = "K = 1")
plt.plot(np.arange(0, 8, .01), y_est2)
plt.grid()
plt.show()
```
